```python
import random
def print_original_plot(name,path,outlier_delete=False):
  material=path+name
  filename="/all_result.csv"
  df=pd.read_csv(material+filename)
  df=df.dropna(axis=0) # drop the unconsistent data long
  df=df.drop(list(df)[0],axis=1)
  # print(df.describe())
  if outlier_delete:
    randomlist = random.sample(range(0, len(df)), 1000)
    for i in randomlist:
      mean = df.iloc[i].mean()
      std = df.iloc[i].std()

      z_scores = (df.iloc[i] - mean) / std

      mask = np.abs(z_scores) > 5
      # print(mask)
      drop_col=[]
      for i in range(len(mask)):
        if mask[i]:
          drop_col.append(df.columns[i])
      df.drop(drop_col,axis=1, inplace=True)
  return df
```

```python
def fft_self(df):
    data_rows = df.iloc[:-1, :]  # All rows except the last
    label_row = df.iloc[-1, :]    # Last row (label)

    # Apply FFT on each column except the last row
    fft_result = data_rows.apply(lambda col: np.abs(np.fft.fft(col)), axis=0)

    # Append the label row to the FFT result
    fft_result.loc['label'] = label_row
    return fft_result
```

```python
def find_distants(miss_label,targetdata,plot_target,real):
    # miss_label=0
    ground_truth=dis_table[dis_table['label'] == miss_label].drop('label',axis=1)
    # print(targetdata.shape,ground_truth.shape)
    # targetdata=ground_truth[0].values
    distants=[float(np.linalg.norm(targetdata-(ground_truth.iloc[i].values))) for i in range(len(ground_truth)) ]
    # Find the index of the minimum distance
    min_dist_index = np.argmin(distants)
    # Get the closest data point from table
    closest_data = plot_table.iloc[min_dist_index].values
    plt.figure(figsize=(4, 3))
    plt.plot(plot_target, label="Target Data", marker='o', linestyle='-', color='blue')
    if real:
        plt.plot(closest_data, label="Closest Correct Data", marker='x', linestyle='--', color='green',alpha=0.5)
    else:
        plt.plot(closest_data, label="Closest Wrong Data", marker='x', linestyle='--', color='red',alpha=0.5)
    plt.legend()
    plt.xlabel("Features")
    plt.ylabel("Values")
    plt.title(f"Target Data vs Closest Data (Label: {miss_label})")
    plt.show()
    print(f"Minimum Distance: {min(distants)}")
```
