```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np  
# Assuming the result path is stored in a variable called result_path
result_path = './result/Jan_19_overfitting/train/'

file_list = os.listdir(result_path)

loss_values = np.array([])
for file_name in file_list:
    if file_name.endswith('.csv'):
        file = result_path + file_name
        data = pd.read_csv(file)
        loss_values = np.append(loss_values, data['loss'].values)
print(loss_values)
# Plot the loss values
plt.figure(figsize=(15, 6))
plt.plot(loss_values, label='Loss')
plt.xlabel('Backpropagation Iteration')
plt.ylabel('Loss')
plt.title('Loss Value Plot')
plt.legend()
plt.show()
```

```python
result_path = './result/Jan_19_overfitting/train/'

file_list = os.listdir(result_path)

length = np.array([])
for file_name in file_list:
    if file_name.endswith('.csv'):
        file = result_path + file_name
        data = pd.read_csv(file)
        length = np.append(length, len(data))
print(length)
# Plot the loss values
plt.figure(figsize=(10, 6))
plt.plot(length, label='length')
plt.xlabel('Epoch')
plt.ylabel('length ')
plt.title('length for each epoch')
plt.legend()
plt.show()
```

```python
result_path = './result/Jan_19_overfitting/test/'

file_list = os.listdir(result_path)

length = np.array([])
for file_name in file_list:
    if file_name.endswith('.csv'):
        file = result_path + file_name
        data = pd.read_csv(file)
        
        length = np.append(length, len(data))
print(length)
# Plot the loss values
plt.figure(figsize=(10, 6))
plt.plot(length, label='length')
plt.xlabel('Epoch')
plt.ylabel('length ')
plt.title('length for each epoch')
plt.legend()
plt.show()
```

```python
result_path = './result/Jan_19_overfitting/test/'

file_list = os.listdir(result_path)

rewards = np.array([])
for file_name in file_list:
    if file_name.endswith('.csv'):
        file = result_path + file_name
        data = pd.read_csv(file)
        
        rewards = np.append(rewards, data.iloc[len(data)-1]['tot_reward'])
print(rewards)
# Plot the loss values
plt.figure(figsize=(10, 6))
plt.plot(rewards, label='rewards')
plt.xlabel('Epoch')
plt.ylabel('rewards ')
plt.title('rewards for each epoch')
plt.legend()
plt.show()
```
